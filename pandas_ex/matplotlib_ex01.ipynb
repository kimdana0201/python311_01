### Tick_patames()함수: 눈금의 스타일을 다양하게 설정

- axis: 설정이 적용될 축을 지정
- direction:'in', 'out'으로 설정하면 눈금이 안/밖으로 표시, {'in', 'out', 'inout'} 중 선택
- length: 눈금의 길이를 지정
- pad: 눈금과 레이블과의 거리를 지정
- labelsize: 레이블의 크기를 지정
- labelcolor: 레이블의 색상을 지정
- top/bottom/left/right: True/False로 지정하면 눈금이 표시될 위치를 선택
- width : 눈금의 너비를 지정
- color: 눈금의 색상을 지정

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0., 5., 0.2)
plt.title("Line Plot  표")
plt.plot(t, t, 'r--', t, 0.5 * t**2, 'bs:', t, 0.2 * t**3, 'g^-')
plt.legend(['Red', 'Blue', 'Green'], loc='best', ncol=2, fontsize=11, frameon=True, shadow=True)  # ncol 열 개수 지정
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

mpl.rcParams['axes.unicode_minus'] = False
x1 = np.linspace(start=0.0, stop=5.0, num=50)
x2 = np.linspace(0.0, 2.0, num=50)
y1 = np.cos(2 * np.pi * x1) * np.exp(-x1)
y2 = np.cos(2 * np.pi * x2)

plt.suptitle("Main Title")
plt.subplot(2, 1, 1) #2행 1열 1개 만드는 구조
plt.plot(x1, y1, 'yo-')
plt.title('A tale of 2-1 subplots') #이 타이틀만 해당.
plt.xlabel('time (m)')
plt.ylabel('Damped oscillation')

plt.subplot(2, 1, 2)
plt.plot(x2, y2, 'r.-')
plt.xlabel('time (s)')
plt.title('A tale of 2-2 subplots')
plt.ylabel('Undamped')

plt.tight_layout()
plt.show()
```

### 왜 다 똑같이 출력이 될까?
- np.random.seed(0) -> 랜덤 x, 재현 가능 랜덤. 이걸 설정하면 난수가 진짜 랜덤이 아닌, 항상 같은 순서로 재현되는 값이 나온다.
  seed는 난수 생성기의 시작점(초깃값)을 고정하는 역할. 0으로 고정했기 때문에 실행할 때마다: x, y1, y2, y3 전부 동일한 값이 생성됨.
  결과적으로 그래프도 항상 똑같이 나옴.
  실험 재현, 디버깅, 논문/과제 결과 일치에서 많이 쓰임 (일부러 동일하게 만들고 싶을 때)

```python
import matplotlib.pyplot as plt
import numpy as np

def f(x,y):
    return(1-x/2+x**5+y**3) * np.exp(-x ** 2 - y**2)  #수치를 부여하는 함수임 너무 의미 주지 마!
n = 256
x = np.linspace(-3, 3, n) #256행
y = np.linspace(-3, 3, n) #256열
XX, YY = np.meshgrid(x, y)
ZZ = f(XX, YY)
plt.title("Contour plots")
plt.contourf(XX, YY, ZZ, alpha=.75, cmap='hot')
plt.contour(XX, YY, ZZ, colors='black')
```

```python
import matplotlib.pyplot as plt
import numpy as np

y1 = [70, 90, 80]
y2 = [60, 70, 90]
x = np.arange(len(y1))
err1 = np.abs(np.random.randn(3)) + 5
err2 = np.abs(np.random.randn(3)) + 5
xlabel=['국어', '영어', '수학']
plt.title("Barh Chart")
plt.barh(x, y1, height=0.3, xerr=err1, ecolor='Yellow', color='red', align='center', label='점수')
plt.barh(x+0.3, y2, height=0.3, xerr=err2, color='blue', align='center', label='반평균') # x로부터 0.4 만큼 떨어진 곳에 해라.
plt.yticks(x+0.3/2, ylabel)
plt.xlim(0, 100)
plt.ylabel("과목")
plt.xlabel("점수")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.rc('font', family = "Malgun Gothic")
plt.rc('font', size = 12)
plt.rc('axes', unicode_minus = False)

year = [2018, 2019, 2020, 2021, 2022]
physics = [6.2, 6.3, 6.0, 5.9, 6.8]
chemistry = [10.6,9.4, 8.0, 8.0, 8.0]
life = [16.1, 17.7, 16.1, 13.1, 14.7]
earth = [16.7, 17.7, 16.1, 13.0, 14.9]

science = {
    "year": year,
    "physics": physics,
    "chemistry": chemistry,
    "life" : life,
    "earth": earth
}
sci = pd.DataFrame(science)
sci
```

```python
import matplotlib.pyplot as plt

labels = ['개구리', '돼지', '개', '통나무']
sizes = [15.55, 30, 45, 9.45]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)
plt.title("Pie Chart")
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.2f%%', shadow=True,
        startangle=90, counterclock=False) #시계방향
plt.axis('equal')
plt.show()
```
