```python
import pandas as pd
import numpy as np
```

```python
s1 = pd.Series([0,1] , index = ['A', 'B'])
s1
```

```python
s2 = pd.Series([2, 3, 4], index=['A', 'B', 'B'])
s2
```

```python
s3 = pd.concat([s1, s2])
s3
```

```python
s3 = pd.concat([s1, s2], axis=1)
s3
```

```python
# arange() : python의 range() 함수와 동일하게 동작.
df1 =  pd.DataFrame(np.arange(6).reshape(3,2),
                    index=['a', 'b', 'c'],
                    columns=['데이터1', '데이터2'])
df1
```

```python
df2 = pd.DataFrame(5 + np.arange(4).reshape(2,2),
                   index=['a', 'c'],
                   columns=['데이터3', '데이터4'])
df2
```

```python
df3 = pd.concat([df1,df2], axis=1)
df3
```

```python
df3 = pd.concat([df1,df2], axis=0)
df3
```

# join()

```python
df1 = pd.DataFrame([[1., 2], [3., 4.], [5., 6.]],
                   index=['a', 'c', 'e'],
                   columns=['서울', '부산'])
df1
```

```python
df2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13., 14.]],
                   index=['b', 'c', 'd', 'e'],
                   columns=['대구', '광주'])
df2
```

```python
# join() : 인덱스를 기준으로 병합
df3 = df1.join(df2, how='outer')
df3
```

```python
df4 = df1.join(df2, how='inner')
df4
```

```python
df5 = df1.join(df2, how='left')
df5
```

```python
df6 = df1.join(df2, how='right') 
df6
```

```python
n = np.nan
idx = ['row1', 'row2', 'row3']
col = ['col1', 'col2', 'col3']
data = [[1,2,200], [100,5,6],[7,300,n]]
df = pd.DataFrame(data, idx, col)
df
```

```python
# idxmax(): 최댓값이 포함되어있는 행 또는 열을 구할 때 사용
df.idxmax(axis=0)
```

```python
df.idxmax(axis=1)
```
