### pivot_table
형식) pivot_table(data, values=None, index=None, columns=None, aggfunc='mean', fill_value=None, margins=False, margins_name='All')   값이 여러 개 있다면 집계함수가 필요하다
- data: 분석할 데이터프레임 (메서드일 때는 필요하지 않음)
- values: 분석할 데이터프레임에서 분석할 열
- index: 행 인덱스로 들어갈 키 열 또는 키 열의 리스트
- columns: 열 인덱스로 들어갈 키 열 또는 키 열의 리스트
- aggfunc: 분석 메서드 지정
- fill_value: NaN 대체 값 설정
- margins: 오른쪽과 아래에 합계를 붙일지 여부 지정
- margins_name: 합계 열(행)의 이름 지정

```python
##pd.pivot_table(data, index=, columns=, values=, aggfunx=)
index -> 행 기준
columns -> 열 기준
values -> 계산할 값
aggfunc -> 계산 방법
```

```python
df2 = pd.DataFrame({'key1': ['A', 'A', 'B', 'B', 'A'],
                    'key2': ['one', 'two', 'one', 'two', 'one'],
                    'data1': [1,2,3,4,5],
                    'data2': [10, 20, 30, 40, 50] })
df2
```

```python
df2.pivot_table("data1", "key1", "key2")  #key1 = 행인덱스와 key2 열 인덱스로 만들고 
# default 로 aggfunc='mean'
```

```python
df2.pivot_table("data1", "key1", "key2", aggfunc = 'sum')  #aggfunc는 집계방법이다.
```

```python
df2.pivot_table("data1", "key1", "key2", margins = True)
```

```python
print(df2.pivot_table("data1", "key1", "key2", margins=True, margins_name='평균'))
```

```python
df2.pivot_table("data1", index=["key1", "key2"])
```

```python
df2.pivot_table("data1", index=["key1", "key2"])
```

### 그래프 조정
 - plt.figure(figsize=(10,5))  그래프 크기 키우는 방법
 - plt.ylim(0,8)는 y축 키우는 방법
 - plt.xlim(0,20)는 x축 키우는 방법
 - plt.title("Iris 데이터의 Bar Plot", color = 'blue') title에 컬러를 지정할 수 있음
 - iris[:5].plot.bar(rot=0, color='skyblue') 막대 그래프의 색 변경 방법
   - x/y축 글자 색 변경
       - plt.xlabel("Data", color='red')
       - plt.ylabel("각 Feature의 값", color='green')
    - 축 눈금 글자 색 변경
       - plt.wticks(color='purple')
       - p;t.yticks(color='orange')
#### 그래프 종류
 - df.plot(kind='line') or df.plot()  --> 시간에 따른 매출 변화
 - df.plot(kind='bar') --> 막대 그래프, 값 비교할 때 가장 많이 사용된다
    - df.plot(kind='barh') -->  가로 막대 그래프
 - df.plot(kind='hist') ---> 시험 점수 분포
 - df.plot(kind='box') ---> 박스 플롯(box) .. 최소값, 최대값, 중앙값, 이상치
 - df.plot(kind='scatter', x='A', y='B') ---> 두 변수 관계 확인 ( 키 VS 몸무게 관계)
 - df.plot(kind='pie')
