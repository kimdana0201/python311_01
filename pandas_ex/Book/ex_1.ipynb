```python
def remouve_comma(x):
    print(x, 'in function')
    return x
```

```python
s = Series(['1,234', '5,678','9,876'])
result = list(s.map(remouve_comma))  #map함수는 반복자 객체가 앞에 있다. 데이터 개수만큼 반복한다. 3개 있으니 3번 반복 호출한다.
print(result)
```

```python
def remove_comma(x) :
    return int(x.replace(",", ""))
```

```python
s = Series(['1,234', '5,678','9,876'])
result = list(s.map(remouve_comma)) 
print(result)             #""없애기 근데 출력이 안됨 다ㅣㅅ 확인
```

```python
from pandas import Series

data = [3.1, 2.0, 10.1, 5.1]
index = ['000010', '000020', '000030', '000040']
s = Series(data=data, index=index)
print(s)
```
