```python
# [문제2] 아래 코드를 활용해서 주어진 문제에 대한 코드를 작성하시오.
data = [
    ["2차전지(생산)", "SK이노베이션", 10.19, 1.29],
    ["해운", "팬오션", 21.23, 0.95],
    ["시스템반도체", "티엘아이", 35.97, 1.12],
    ["해운", "HMM", 21.52, 3.20],
    ["시스템반도체", "아이에이", 37.32, 3.55],
    ["2차전지(생산)", "LG화학", 83.06, 3.75]
]

columns = ["테마", "종목명", "PER", "PBR"]
df = pd.DataFrame(data=data, columns=columns)
df
```

```python
#(1) 데이터프레임에서 테마별 PER, PBR의 평균을 계산하세요.
df.groupby('테마')[['PER', 'PBR']].mean()
```

```python
#(2) 데이터프레임에서 테마별 종목의 개수를 계산하세요.
df.groupby(['테마']).count()
```

```python
# (3) 데이터프레임에서 테마별 PER의 평균을 계산하세요.
df.groupby('테마')[['PER']].mean()
```

```python
# (4) 데이터프레임에 대해 테마로 그룹화한 후 PER, PBR 컬럼을 슬라이싱하세요. 
# 그 다음 두 컬럼에 대한 평균을 계산하세요.

df2 = df.loc[:, "PER": "PBR"].groupby(df['테마']).mean()
df2
```

```python
#(5) 데이터프레임에 대해서 '테마' 컬럼으로 그룹화한 후 
#PER 컬럼은 max, PBR 컬럼은 min 함수를 적용해보세요.
df3 = df.groupby('테마').agg({
    'PER': 'max',
    'PBR': 'min'
})
df3
```
