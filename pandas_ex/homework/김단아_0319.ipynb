```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./titanic_train.csv')
df
```

```python
#[문제1] Survived 컬럼의 생존자와 사망자 수를 구하여 막대 그래프로 나타내시오.
import pandas as pd
import matplotlib.pyplot as plt

plt.rc('font', family='Malgun Gothic')
plt.rc('axes', unicode_minus=False)

counts = df['Survived'].value_counts().sort_index()

ax = counts.plot.bar(color=['black','red'], rot=0)

plt.xticks([0, 1], ['사망자', '생존자'])
plt.ylabel('총 탑승객 수')
plt.xlabel('생존 여부')
plt.title('타이타닉 탑승객 생존 여부')

for i, v in enumerate(counts):
    ax.text(i, v + 5, str(v)+"명", ha='center')

plt.show()
```

```python
#(2) Survived 컬럼의 생존자와 사망자 수를 등급별로 구하여 누적막대그래프로 나타내시오. 등급을 누적으로 나타낸다.
import matplotlib.pyplot as plt
import pandas as pd

plt.rc('font', family = "Malgun Gothic")
plt.rc('font', size = 12)
plt.rc('axes', unicode_minus = False)

total = df.groupby(['Pclass', 'Survived'])['Survived'].size()
result = pd.DataFrame(total).unstack()
color= ['black', 'red']
result.plot(kind = 'bar', stacked=True, rot=0)
plt.show()
```

```python
# (3) Embarked 컬럼의 탑승지역의 분포를 파이차트로 나타내는데 각 탑승지역에 해당하는 비율을 파이차트에 출력하시오.
import pandas as pd
import matplotlib.pyplot as plt

df2 = df['Embarked'].value_counts()
df2.plot.pie(autopct='%.2f%%')

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False

plt.title('탑승지역의 분포')
plt.show()
```
