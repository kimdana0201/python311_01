```python
#[문제] 아래 코드를 참고하여 주어진 문제에 대한 코드를 완성하시오.

import matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset("iris")
iris
```

```python
#(1) iris 데이터에서 species별 sepal_length의 평균을 구한 후 이를 'bar' 타입으로 그려보세요.
import matplotlib.pyplot as plt
iris1 = iris.groupby('species')[['sepal_length']].mean()
iris1.plot(kind='bar', rot=0)
```

```python
# (2) iris 데이터에서 품좀별 데이터의 개수를 파이 차트 형태로 그려보세요.

iris2 = iris['species'].value_counts()
iris2.plot.pie(autopct='%.2f%%')

plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus']=False

plt.title('품종별 데이터 개수')
plt.show()
```
